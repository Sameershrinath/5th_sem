```python
import requests
import random
import time
from datetime import datetime

url = "https://gfgcdn.com/tu/Vb4/"

user_agents = [
    "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0 Safari/537.36",
    "Mozilla/5.0 (Macintosh; Intel Mac OS X 12_5) AppleWebKit/605.1.15 (KHTML, like Gecko) Version/15.4 Safari/605.1.15",
    "Mozilla/5.0 (Linux; Android 12; SM-G991B) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Mobile Safari/537.36",
    "Mozilla/5.0 (iPhone; CPU iPhone OS 16_0 like Mac OS X) AppleWebKit/605.1.15 (KHTML, like Gecko) Mobile/15E148",
]

referers = [
    "https://www.google.com/",
    "https://www.facebook.com/",
    "https://www.linkedin.com/",
    "https://twitter.com/",
    "https://www.instagram.com/",
    "https://github.com/",
]

# Initialize counters
total_requests = 0
successful_requests = 0
failed_requests = 0
start_time = datetime.now()

print(f"🚀 Starting automated requests at {start_time.strftime('%Y-%m-%d %H:%M:%S')}")
print("=" * 60)

for i in range(100):  # Increased number for long-running
    headers = {
        "User-Agent": random.choice(user_agents),
        "Referer": random.choice(referers),
        "Accept-Language": "en-US,en;q=0.9",
    }

    try:
        response = requests.get(url, headers=headers, allow_redirects=True, timeout=10)
        total_requests += 1
        successful_requests += 1
        
        current_time = datetime.now().strftime('%H:%M:%S')
        print(f"[{current_time}] Request #{total_requests} ✅ - Status: {response.status_code}")
        print(f"    📊 Total Sent: {total_requests} | Success: {successful_requests} | Failed: {failed_requests}")
        
    except Exception as e:
        total_requests += 1
        failed_requests += 1
        
        current_time = datetime.now().strftime('%H:%M:%S')
        print(f"[{current_time}] Request #{total_requests} ❌ - Error: {str(e)[:50]}...")
        print(f"    📊 Total Sent: {total_requests} | Success: {successful_requests} | Failed: {failed_requests}")

    # Progress update every 10 requests
    if total_requests % 10 == 0:
        elapsed_time = datetime.now() - start_time
        success_rate = (successful_requests / total_requests) * 100
        print("\n" + "="*60)
        print(f"📈 PROGRESS UPDATE - {total_requests} requests completed")
        print(f"⏱️  Elapsed Time: {str(elapsed_time).split('.')[0]}")
        print(f"✅ Success Rate: {success_rate:.1f}%")
        print("="*60 + "\n")

    # Random delay between requests
    delay = random.uniform(3, 8)
    print(f"    ⏳ Waiting {delay:.1f}s before next request...\n")
    time.sleep(delay)

# Final summary
end_time = datetime.now()
total_time = end_time - start_time
success_rate = (successful_requests / total_requests) * 100 if total_requests > 0 else 0

print("\n" + "🏁 FINAL SUMMARY " + "="*45)
print(f"📊 Total Requests Sent: {total_requests}")
print(f"✅ Successful: {successful_requests}")
print(f"❌ Failed: {failed_requests}")
print(f"📈 Success Rate: {success_rate:.1f}%")
print(f"⏱️  Total Time: {str(total_time).split('.')[0]}")
print(f"🕐 Started: {start_time.strftime('%Y-%m-%d %H:%M:%S')}")
print(f"🕐 Ended: {end_time.strftime('%Y-%m-%d %H:%M:%S')}")
print("="*60)
```

```python
import requests
import random
import time

url = "https://gfgcdn.com/tu/Vb4/"

# List of user-agents (real desktop + mobile devices)
user_agents = [
    "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 Chrome/115.0 Safari/537.36",
    "Mozilla/5.0 (Macintosh; Intel Mac OS X 13_3) AppleWebKit/605.1.15 Version/16.3 Safari/605.1.15",
    "Mozilla/5.0 (iPhone; CPU iPhone OS 16_0 like Mac OS X) AppleWebKit/605.1.15 Mobile/15E148",
    "Mozilla/5.0 (Linux; Android 11; SM-A515F) AppleWebKit/537.36 Chrome/111.0 Mobile Safari/537.36",
    "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 Chrome/113.0 Safari/537.36"
]

# Referers to simulate traffic from different platforms
referers = [
    "https://www.google.com/",
    "https://www.facebook.com/",
    "https://www.linkedin.com/",
    "https://twitter.com/",
    "https://www.instagram.com/"
]

# Function to simulate realistic user traffic
def simulate_clicks(count=50):
    for i in range(count):
        headers = {
            "User-Agent": random.choice(user_agents),
            "Referer": random.choice(referers),
            "Accept-Language": "en-US,en;q=0.9",
            "Connection": "keep-alive",
        }

        try:
            response = requests.get(url, headers=headers, allow_redirects=True, timeout=10)
            print(f"[{i+1}] Status: {response.status_code}, URL: {response.url}")
        except Exception as e:
            print(f"[{i+1}] Error:", e)

        # Random delay between 5 to 15 seconds
        time.sleep(random.uniform(5, 15))

simulate_clicks(count=30)  # Start with 30 safe clicks
```
