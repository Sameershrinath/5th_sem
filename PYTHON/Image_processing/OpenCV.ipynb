```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import cv2
```

```python
from glob import glob # this is for reading the image from folder
```

```python
graphs=glob("../Data_Science/Dataset/*.png")
```

```python
mpl_image=plt.imread(graphs[0])
cv2_image=cv2.imread(graphs[0])
```

Displaying the image

```python
fig,ax=plt.subplots(figsize=(2,2))
ax.imshow(mpl_image)
ax.axis("off")
plt.show()
```

```python
plt.style.available
```

```python
plt.style.use("Solarize_Light2")

fig,axs=plt.subplots(1,4,figsize=(15,5))
axs[0].imshow(mpl_image[:,:,0],cmap="Reds")
axs[1].imshow(mpl_image[:,:,1],cmap="Greens")
axs[2].imshow(mpl_image[:,:,2],cmap="Blues")
axs[3].imshow(mpl_image,cmap="Blues")
plt.show()
```

```python
fig,axs=plt.subplots(1,4,figsize=(15,5))
axs[0].imshow(cv2_image[:,:,0],cmap="Blues")
axs[1].imshow(cv2_image[:,:,1],cmap="Greens")
axs[2].imshow(cv2_image[:,:,2],cmap="Reds")
axs[3].imshow(cv2_image,cmap="Blues")
plt.show()
```

```python
cv2_image_RGB=cv2.cvtColor(cv2_image, cv2.COLOR_BGR2RGB)
```

```python
fig,axs=plt.subplots(1,4,figsize=(15,5))
axs[0].imshow(cv2_image_RGB[:,:,0],cmap="Reds")
axs[1].imshow(cv2_image_RGB[:,:,1],cmap="Greens")
axs[2].imshow(cv2_image_RGB[:,:,2],cmap="Blues")
axs[3].imshow(cv2_image_RGB,cmap="Blues")
plt.show()
```

```python
cv2_image_grey=cv2.cvtColor(cv2_image, cv2.COLOR_BGR2GRAY)

fig,axs=plt.subplots(figsize=(15,5))

axs.imshow(cv2_image_grey,cmap="grey")
plt.show()
```
