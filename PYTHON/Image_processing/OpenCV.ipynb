we can also use 

```python
plt.imshow(cv2_image_grey, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# Convert cv2_image to grayscale using matrix logic (manual conversion)
# The standard formula: Gray = 0.299*R + 0.587*G + 0.114*B
# Grayscale = (R + G + B) / 3

# cv2_image is in BGR format
B = cv2_image[:, :, 0]
G = cv2_image[:, :, 1]
R = cv2_image[:, :, 2]

cv2_image_grey_manual = (0.114 * B + 0.587 * G + 0.299 * R).astype(np.uint8)

plt.imshow(cv2_image_grey_manual, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# Convert cv2_image to grayscale using the average method
cv2_image_grey_avg = ((R.astype(np.float32) + G.astype(np.float32) + B.astype(np.float32)) / 3).astype(np.uint8)

plt.imshow(cv2_image_grey_avg, cmap='gray')
plt.axis('off')
plt.show()
```

## **resizing the image**

```python
img_resized=cv2.resize(cv2_image_RGB,(5,5),interpolation=cv2.INTER_CUBIC)

fig,ax=plt.subplots(figsize=(4,4))
ax.imshow(img_resized)
ax.axis("on")
plt.show()
```

## **Sharpening the image**

```python
kerner_sharp=np.array([[-1,-1,-1],
                       [-1,10,-1],
                       [-1,-1,-1]])
sharpened_img=cv2.filter2D(cv2_image_RGB,-1,kerner_sharp)


fig,ax=plt.subplots(figsize=(4,4))
ax.imshow(sharpened_img)
ax.axis("off")
plt.show()
```

```python
kerner_sharp=np.array([[-1,-1,-1],
                       [-1,10,-1],
                       [-1,-1,-1]])
sharpened_img=cv2.filter2D(cv2_image_RGB,-1,kerner_sharp)


fig,ax=plt.subplots(figsize=(4,4))
ax.imshow(sharpened_img)
ax.axis("off")
plt.show()
```

## **Blurring the image**

```python
kernel_blurring= np.ones((5,5),np.float32)/25

blurred_img=cv2.filter2D(cv2_image_RGB,-1,kernel_blurring)

fig,ax=plt.subplots(figsize=(4,4))
ax.imshow(blurred_img)
ax.axis("off")
plt.show()
```
