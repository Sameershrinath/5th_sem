Image processing in real time using webcam

```python
import cv2 as cv
import numpy as np 
import matplotlib.pyplot as plt
```

Working with webcam

```python
# openign the normal camera

cam=cv.VideoCapture(0)

while True:
    _,img=cam.read()
    img=cv.flip(img,1)
    cv.imshow("camera",img)

    if cv.waitKey(1)==27:
        cam.release()
        cv.destroyAllWindows()
        break
```

```python
plt.imshow(img)
```

```python
#mirroring the image
img=cv.flip(img,)
plt.imshow(img)
```

```python
#mirroring the image
img=cv.flip(img,0)
plt.imshow(img)
```
