Gradient of RGB

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

Red gradient

```python
img=[]
for i in range(256):
    t=[]
    for j in range(256):
        t.append([i,0,0])
    img.append(t)
for i in range(256,0,-1):
    t=[]
    for j in range(256,0,-1):
        t.append([i,0,0])
    img.append(t)
        
```

```python
img=np.array(img)
```

```python
img.shape
```

```python
plt.imshow(img)
```

Green gradient

```python
img=[]
for i in range(256):
    t=[]
    for j in range(700):
        t.append([0,i,0])
    img.append(t)
for i in range(256,0,-1):
    t=[]
    for j in range(700,0,-1):
        t.append([0,i,0])
    img.append(t)
        
```

```python
plt.imshow(img)
```

Blue gradient

```python
img=[]
for i in range(256):
    t=[]
    for j in range(700):
        t.append([0,0,i])
    img.append(t)
for i in range(256,0,-1):
    t=[]
    for j in range(700,0,-1):
        t.append([0,0,i])
    img.append(t)
        
```

```python
plt.imshow(img)
```

combining all the gradient

```python
img=[]
for i in range(256):
    t=[]
    for j in range(700):
        t.append([i,0,0])
    img.append(t)
for i in range(255,0,-1):
    t=[]
    for j in range(700,0,-1):
        t.append([i,0,0])
    img.append(t)

for i in range(256):
    t=[]
    for j in range(700):
        t.append([0,i,0])
    img.append(t)
for i in range(255,0,-1):
    t=[]
    for j in range(700,0,-1):
        t.append([0,i,0])
    img.append(t)
        
for i in range(256):
    t=[]
    for j in range(700):
        t.append([0,0,i])
    img.append(t)
for i in range(255,0,-1):
    t=[]
    for j in range(700,0,-1):
        t.append([0,0,i])
    img.append(t)
img=np.array(img,dtype=np.uint8)
        
```

```python
plt.imshow(img)
```

```python
cv.imwrite("gradient.png",img)
```
