# Dealing with the duplicated rows in the data set

```python
import pandas as pd 
df= pd.read_csv("dataset/AB_NYC_2019.csv")
```

```python
df.head()
```

```python
df.duplicated().sum()
```

```python
df[df.duplicated()]
```

```python
df[df.duplicated(keep='first')]
```

```python
df[df.duplicated(keep='last')]
```

```python
df[df.duplicated(keep=False)]
```

```python
df.drop_duplicates(keep="first",inplace=True)
```

```python
df.duplicated().sum()
```

```python
# Finallly we have deleted all the duplicated values from our data set 
# functions duplicated() drop_duplicates() 
# keep= first to keep teh first row 
# keep = last to keep the last row 
```
