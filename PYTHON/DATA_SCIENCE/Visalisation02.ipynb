# **Matplotlib**

- %matplotlib inline   : for the inline static plot
- %matplotlib notebook : for the interactive graph

```python
import numpy as np 
import matplotlib as mp
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=np.linspace(0,np.pi*10,100)
```

```python
cos_y=np.cos(x)
plt.plot(x,cos_y)
plt.title("Cos")
plt.show()
```

```python
cos_y=np.cos(x)
sin_y=np.sin(x)

plt.plot(x,sin_y,label="Sine")
plt.plot(x,cos_y,label="Cosine")
plt.legend(loc="upper right")
plt.title("Sin and cos plot")
plt.show()
```

```python
x=np.linspace(0,np.pi*4,400)
sin_y=np.sin(x)
cos_y=np.cos(x)
plt.plot(x,sin_y,"*")
plt.plot(x,cos_y,"--")
plt.title("Sin and cos plot")
plt.show()
```

### **Setting styles to the graph**

```python
x=np.linspace(0,np.pi*10,100)
sin_y=np.sin(x)
cos_y=np.cos(x)

li=plt.style.available
for i in li:
    plt.figure(figsize=(18,5))
    plt.style.use(i)
    plt.plot(x, sin_y, label="Sine")
    plt.plot(x, cos_y, label="Cosine")
    plt.title(f"Type:{i}")
    plt.legend()
    plt.savefig(f"plot_style_{i}.png")
    plt.show()
```

## **Different types of plots**

```python
x=np.linspace(0,np.pi*4,400)
sin_y=np.sin(x)
plt.style.use("Solarize_Light2")
plt.ylim(-5,5)
plt.plot(x,sin_y)
plt.show()
```

We can use set function to pass different parameters in one line

```python
# Scatter Plot
x1=np.random.randn(500)
y1=np.random.randn(500)

plt.scatter(x1,y1)
plt.show()
```

```python
# Scatter Plot
x1=np.random.randn(5000)
y1=np.random.randn(5000)

x2=np.random.randn(5000)+2
y2=np.random.randn(5000)+2

plt.scatter(x1,y1,c="red",alpha=0.5,label="Group 1")
plt.scatter(x2,y2,c="yellow",alpha=0.8,label="Group 2")
plt.legend()
plt.show()
```

```python
rng=np.random.RandomState(0)

x=np.random.randn(100)
y=np.random.randn(100)

color=rng.rand(100)
sizes=rng.rand(100)*500

plt.scatter(x,y,c=color,s=sizes,alpha=.7,cmap='viridis',edgecolors="black")
plt.show()
```
