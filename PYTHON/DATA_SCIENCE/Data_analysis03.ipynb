## **Heart data (EDA)**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("./Dataset/heart.csv")
```

```python
df.head()
```

```python
df.isna().sum(),df.shape
# we have not any null value
```

```python
df.nunique()
```

Correlation

```python
sns.heatmap(df.corr(numeric_only=True),cmap="Greens")
```

Distplot of the age

```python
import warnings
warnings.filterwarnings("ignore")

fig, axes = plt.subplots(1, 2, figsize=(16, 4))

sns.boxenplot(data=df, x="Age", ax=axes[0])
axes[0].set_title("Boxenplot of Age")

sns.histplot(df["Age"], kde=True, ax=axes[1])
axes[1].set_title("Distribution of Age")

plt.tight_layout()
plt.show()
```
